```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pylab import figure, setp, axes
import matplotlib.patches as patches

def setBoxColors(bp):
    params = ["boxes", "caps", "whiskers", "medians"]
    colors = ["red", "green", "blue", "orange"]
    for param in params:
        for i in range(len(colors)):
            if param in ["whiskers", "caps"]:
                setp(bp[param][i*2], color=colors[i])
                setp(bp[param][i*2+1], color=colors[i])
            else:
                setp(bp[param][i], color=colors[i])
            

fig = figure(figsize=(33, 3))
ax = axes()

data = {}
with open('../training/results/importance_mawi_benign_mawi_ddos.json') as f:
    data = json.load(f)


ml_algo_names = {
    "RandomForestClassifier": "Random Forest",
    "SVC": "SVM",
    "DecisionTreeClassifier": "Decision Tree",
    "MLPClassifier": "MLP"
}
ml_algos = list(data.keys())
features = list(map(lambda x: str(x), list(data[ml_algos[0]].keys())))
metrics = list(data[ml_algos[0]][features[0]].keys())

print(features)
print(metrics)

metric = "accuracy"

xs = np.arange(len(features))
feature_metric_map = {}
for feature in features:
    feature_metric_map[feature] = []
    for ml_algo in ml_algos:
        feature_metric_map[feature].append(data[ml_algo][feature][metric])


for feature in feature_metric_map:
    p = xs[features.index(feature)]
    p = p*4+p
    bp = plt.boxplot(feature_metric_map[feature], positions=[p, p+1, p+2, p+3], widths=0.6, showfliers=False)
    setBoxColors(bp)

# Creating axes instance
ax.set_ylabel(f"Impact on {metric}", fontsize=18)
ax.set_yticks([0.0, 0.1, 0.3, 0.5], ["0%", "10%", "30%","50%"], fontsize=14)
ax.set_xticks(xs*4+xs+1.5, features, rotation=10)
ax.legend([bp["boxes"][0], bp["boxes"][1], bp["boxes"][2], bp["boxes"][3]], [ml_algo_names[ml_algos[0]], ml_algo_names[ml_algos[1]], ml_algo_names[ml_algos[2]], ml_algo_names[ml_algos[3]]], loc="upper left", fontsize=18, ncol=4)
ax.grid(True, linestyle='--', which='major', color='grey')
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_ylim(-0.025, 0.6)

# Create a Rectangle patch
rect = patches.Rectangle((24, -0.02), 5, 0.57, linewidth=2, edgecolor='k', facecolor='none', linestyle='dashed')
# Add the patch to the Axes
ax = plt.gca()
ax.add_patch(rect)

# show plot
import datetime
current_time = datetime.datetime.now().strftime("%Y-%m-%d_%H:%M:%S")
plt.savefig(f"figures/importance_mawi_benign_mawi_ddos_{current_time}.png", bbox_inches='tight', dpi=300)
plt.show()
 
```
